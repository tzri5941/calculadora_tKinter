```python
from tkinter import *


janela = Tk()

janela.geometry("300x350")

janela.resizable(False, False)



def btn_press(x):
    if x == 0:
        txt.insert(END,"0")
    elif x == 1:
        txt.insert(END, "1")
    elif x == 2:
        txt.insert(END, "2")
    elif x == 3:
        txt.insert(END, "3")
    elif x == 4:
        txt.insert(END, "4")
    elif x == 5:
        txt.insert(END, "5")
    elif x == 6:
        txt.insert(END, "6")
    elif x == 7:
        txt.insert(END, "7")
    elif x == 8:
        txt.insert(END, "8")
    elif x == 9:
        txt.insert(END, "9")
    elif x == "+":
        txt.insert(END, "+")
    elif x == "-":
        txt.insert(END, "-")


btn_7 = Button(text="7", height=2, width=4, font="Arial 15 bold", command= lambda: btn_press(7))
btn_7.place(x=10, y=70)
btn_8 = Button(text="8", height=2, width=4, font="Arial 15 bold", command= lambda: btn_press(8) )
btn_8.place(x=70, y=70)
btn_9 = Button(text="9", height=2, width=4, font="Arial 15 bold", command= lambda: btn_press(9))
btn_9.place(x=130, y=70)
btn_4 = Button(text="4", height=2, width=4, font="Arial 15 bold", command= lambda: btn_press(4))
btn_4.place(x=10, y=140)
btn_5 = Button(text="5", height=2, width=4, font="Arial 15 bold", command= lambda: btn_press(5))
btn_5.place(x=70, y=140)
btn_6 = Button(text="6", height=2, width=4, font="Arial 15 bold", command= lambda: btn_press(6))
btn_6.place(x=130, y=140)
btn_1 = Button(text="1", height=2, width=4, font="Arial 15 bold", command= lambda: btn_press(1))
btn_1.place(x=10, y=210)
btn_2 = Button(text="2", height=2, width=4, font="Arial 15 bold", command= lambda: btn_press(2))
btn_2.place(x=70, y=210)
btn_3 = Button(text="3", height=2, width=4, font="Arial 15 bold", command= lambda: btn_press(3))
btn_3.place(x=130, y=210)
btn_0 = Button(text="0", height=2, width=4, font="Arial 15 bold", command= lambda: btn_press(0))
btn_0.place(x=70, y=280)



btn_add = Button(text="+", height=2, width=4, font="Arial 15 bold", command= lambda: btn_press("+"), fg="white", background="black")
btn_add.place(x=230, y=140)

btn_sub = Button(text="-", height=2, width=4, font="Arial 15 bold", command= lambda: btn_press("-"), fg="white",background="black")
btn_sub.place(x=230, y=210)


def limpar():
    txt.delete("0.0","3.0")


btn_del = Button(text="C", height=2, width=4, font="Arial 15 bold", command=limpar, fg="red", background="black")
btn_del.place(x=230, y=280)


def soma():
    campo = txt.get("0.0","3.0")
    indice = campo.find("+")
    n1 = campo[indice+1:]
    n2 = campo[:indice]
    txt.delete("0.0","3.0")
    txt.insert(END,str(int(n1) + int(n2)))

def sub():
    campo = txt.get("0.0","3.0")
    indice = campo.find("-")
    n1 = campo[indice+1:]
    n2 = campo[:indice]
    txt.delete("0.0","3.0")
    txt.insert(END,str(int(n2) - int(n1)))


def igual():
    campo = txt.get("0.0","3.0")
    if campo.find("-") == -1:
        soma()
    else:
        sub()



btn_igual = Button(text="=", height=2, width=4, font="Arial 15 bold", command=igual, fg="green", background="black")
btn_igual.place(x=230, y=70)


txt = Text(height=2, width=21, font="Arial 17 bold")
txt.place(x=10, y=10)






janela.mainloop()



```
